```python
from haversine import haversine
```

```python
start = (float(df.iloc[1]['위도']), float(df.iloc[1]['경도']))
#연세대 37.56586784404183, 126.93859008739398
end = (float(37.56586784404183), float(126.93859008739398))
```

```python
yonsei_point = (float(37.56586784404183), float(126.93859008739398))
```

```python
def get_distance(x) :
    target = (float(x['위도']), float(x['경도']))
    return haversine(target,yonsei_point)
```

```python
get_distance(df.iloc[0])
```

```python
haversine(start,end)
```

```python
#거리 반경 1.3km 설정
```

```python
df['distance'] = df.apply(get_distance, axis=1)
```

```python
def get_total_name(x):
    if x.isnull()['branch_name'] :
        return x['dining_name']
    else :
        if x['branch_name'] in x['dining_name'] :
            return x['dining_name']
        else :
            return x['dining_name'] +' ' + x['branch_name']
```

```python
new_df['total_name'] = new_df.apply(get_total_name, axis=1)
```

```python
new_df = new_df[['total_name','dining_name','branch_name','address','latitude','longitude','distance']]
```

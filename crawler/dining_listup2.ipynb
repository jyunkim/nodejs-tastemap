```python
def make_dining_name(total_name) :
    dining_name = ""
    total_list = total_name.split(" ")
    if '점' in total_list[-1] :
        for data in total_list[:-1]:
            dining_name += data
        return dining_name
    else :
        return total_name
```

```python
def make_branch_name(total_name) :
    branch_name = ""
    total_list = total_name.split(" ")
    if '점' in total_list[-1] :
        branch_name = total_list[-1]
    
    return branch_name
```

```python
df['dining_name'] = df.total_name.map(lambda x : make_dining_name(x))
```

```python
df['branch_name'] = df.total_name.map(lambda x : make_branch_name(x))
```

```python
df = df.rename({"Unnamed: 0" : "dining_id"}, axis=1)
```

```python
df = df[['dining_id','total_name','dining_name','branch_name','category','addr','source']]
```
